```python
#bar plot of the Parental_Education_Level
ax=data['Parental_Education_Level'].value_counts()\
    .head(10).plot(kind='bar',title='Parental_Education_Level',\
                   xlabel='Parental_Education_Level',\
                   ylabel='count',color="green")
```

```python
#histogram of the hours_studied
ax=data['Hours_Studied'].plot(kind="hist",bins=10,title="Hours_Studied")
ax.set_xlabel("Hours_Studied")
ax.set_ylabel("frequency")
```

```python
mean=data['Hours_Studied'].mean()
mean
```

```python
#Density plot
ax=data['Hours_Studied'].plot(kind="kde",title="Hours_Studied")
ax.set_xlabel("Hours_Studied")
ax.set_ylabel("frequency")
plt.axvline(mean,color="yellow")
```

```python
#ploting a scatter plot
data.plot(kind="scatter",x="Hours_Studied",y="Exam_Score",
title="Hours_Studied vs Exam_Score")
plt.show()
```

```python
#plotting using seaborn
sns.scatterplot(x="Hours_Studied",y="Exam_Score",hue="Internet_Access",data=data)
plt.show()
```

```python
#using a pair plot
sns.pairplot(data,vars=["Hours_Studied","Exam_Score","Tutoring_Sessions"],
             hue="School_Type")
plt.show()
```

```python
#correlation
correlation =data[['Exam_Score','Hours_Studied']].corr()
correlation
```
